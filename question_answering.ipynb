```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatOpenAI(
    model="gpt-4o",      # model name from OpenAI
    temperature=0.2     # lower = more deterministic
)
```

```python
from langchain.document_loaders import CSVLoader
loader = CSVLoader(file_path = 'codebasics_faqs.csv', source_column = "prompt", encoding="latin-1")
data = loader.load()
```

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings
from langchain.vectorstores import FAISS

embeddings = HuggingFaceInstructEmbeddings()
vectordb = FAISS.from_documents(documents=data, embedding=embeddings)
```

```python
retriever = vectordb.as_retriever()
retriever.get_relevant_documents("Is knowing Python required?")
```

```python
from langchain.chains import RetrievalQA
chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    input_key="query",
    return_source_documents=True
)
```

```python
chain("should I learn powerbi or tableau?")
```
